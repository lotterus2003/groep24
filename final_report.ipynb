The main objective of this study is to investigate how changes in work behaviour, particularly the rise of hybrid working since the COVID pandemic, have influenced travel behavior and the demand for public transport. The main question is formulated as following: 

**How has the rise of hybrid working since the COVID-19 pandemic influenced the demand for public transport in the Netherlands?** 

To address this main research question, several sub-questions are formulated to guide the analysis. These sub-questions aim to explore how hybrid working has influenced travel behaviour and public transport demand in the Netherlands.

1. Which modes of transport have been affected most from the COVID pandemic? Are there larger changes in certain modes of transport than others? 

2. What would travel demand in the Netherlands have looked like now if there would not have been a COVID crisis? 

3. How has the shift to hybrid working affected peak-hour vs. off-peak demand for public transport?

# Introduction


At the end of 2019 the COVID crisis started and had huge impacts on the daily lives of people in the Netherlands and everywhere else in the world. Due to several lockdowns introduced to limit the number of infections, people were required to keep social distance and avoid unnecessary travel. As a result, many people started working from home instead of commuting to their workplaces. This had major effects on daily travel patterns and the overall use of different modes of transportation.

Public transport was particularly affected. Trains, buses and trams were usually crowded during rush hours but were now nearly empty. The charts below in this chapter will show that travel behaviour did not immediately return to what it was before the crisis. Many people continued to work from home for a part of the week, leading to fewer trips by public transport. These changes are still evident in the use of public transport today and are important for understanding future transport needs. 

This research project looks further into how the use of public transport has changed since the start of the COVID crisis, and what public transport use might have looked like today if the crisis had never happened. In this report, it is assumed that the COVID crisis lasted from 2020 until 2022.
