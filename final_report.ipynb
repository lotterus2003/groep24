## Introduction

At the end of 2019 the COVID-19 pandemic started and had huge impacts on the daily lives of people in the Netherlands and everywhere else in the world. Due to several lockdowns introduced to limit the number of infections, people were required to keep social distance and avoid unnecessary travel. As a result, many people started working from home instead of commuting to their workplaces. Over time, this developed into a more permanent hybrid work culture, in which employees split their working days between home and the office. This shift had major effects on daily travel patterns and the overall use of different modes of transportation.

Graph 1 below shows the changes in the average number of trips per person per year, categorized by purpose of travel. The data include all modes of transport, such as car, train, bicycle, and walking. It is clear that the number of trips sharply decreased after 2019, especially those made for commuting and business purposes. In contrast, trips made for touring or walking slightly increased during the same period. This indicates that although overall mobility declined, people continued to travel locally and for leisure.

Public transport was particularly affected during this period, as trains, buses and trams were usually crowded during rush hours but became nearly empty. Travel behaviour did not immediately return to what it was before the pandemic. Many people continued to work from home for part of the week, leading to fewer commuting trips and lower overall transport demand. Even today, these effects remain visible. Hybrid working has become a structural part of the Dutch labour market, offering employees flexibility to combine home and office work. According to Statistics Netherlands (CBS), over five million people in the Netherlands worked from home sometimes or all of the time in 2023, representing 52 percent of the entire labour force, which is the highest proportion in any EU country (Carter, 2025). While fully remote work has become less common, hybrid working remains dominant and continues to shape national travel behaviour. These structural changes in work patterns are crucial for understanding current and future transport needs.

## 1. Analysis on How Different Modes of Transport Have Been Affected by the COVID-19 Pandemic
To examine how travel behaviour for commuting has evolved since the COVID-19 pandemic, national mobility data from Statistics Netherlands (CBS) was analysed. The dataset contains yearly averages of stages travelled per person per day, average distance per stage, and average travel time, differentiated by age groups, transport motives and transport modes. For this analysis, the motive “Travel to/from work (non-daily commute)” was selected, as it directly reflects changes in commuting behaviour caused by the rise of hybrid working.

Graph 2 presents the average person-kilometres travelled per day for work-related trips across all transport modes combined. Between 2014 and 2019, commuting distances per person remained fairly stable, at around 12-13 km per day. In 2020, however, there is a sharp decline to roughly 7.8 km per day, corresponding with the first year of the COVID-19 pandemic and the introduction of large-scale home-working policies. After 2021, a gradual recovery can be observed, reaching approximately 10 km per day in 2023. This pattern clearly indicates the substantial overall reduction in commuting activity during the pandemic and the partial recovery that followed.

While Graph 2 shows the general trend, it does not yet reveal which transport modes contributed most to this decrease. To address this, Graph 3 disaggregates total commuting distance by transport mode.

Graph 3 shows the average kilometres per person per day by individual transport mode. It becomes immediately clear that the private car (as driver) dominates work-related travel, accounting for the largest share throughout the entire period.

All modes display a visible decline in 2020, but the reduction is most pronounced for public transport modes, particularly train and bus/metro. In contrast, the drop for car travel is more moderate, and the use of active modes such as bike and walking remains relatively stable. The gradual increase after 2021 suggests a partial return to office-based work, though not yet to pre-pandemic levels. To better visualise how the composition of commuting modes changed, graph 4 focuses on relative shares rather than absolute average distances.

Graph 4 displays the percentage share of total work-commute distance per transport mode. Before 2020, the share of car driving accounted for roughly 60% of all person-kilometres. During 2020 and 2021, this share increased sharply to around 68%, while the combined share of public transport (train and bus/metro) decreased notably. This shift implies that people who continued commuting during the pandemic were more likely to use private vehicles rather than public transport, possibly due to health concerns or reduced public transport services.

After 2021, public transport shares recover slightly, but the car remains the dominant mode. To better understand the magnitude of these relative changes per mode, graph 5 compares their developments relative to the pre-COVID baseline year.

This allows a direct comparison of how strongly each mode was affected. Both train and bus/metro show a steep drop in 2020, to around 35-50% of their 2019 levels, and only partial recovery by 2023. Car driving, on the other hand, declines less severely (to around 70% of its 2019 level) and recovers faster, reaching nearly 90% in 2023. This demonstrates that public transport was much more severely impacted by the pandemic and has not yet returned to pre-COVID usage levels, while car commuting has proven more resilient.

## 2. Analysis on How Travel Demand Would Have Evolved Without the COVID-19 Pandemic



To estimate public transport usage in a hypothetical scenario without the impact of COVID-19, a set of assumptions is required. The core assumption is that public transport usage can be predicted based on two key variables: the number of employed individuals and the total population.

The analysis begins by examining the employment trend prior to the COVID-19 pandemic. Using data from 2014 to 2019, the average annual growth in the number of jobs was calculated and extrapolated forward to estimate employment figures for the years 2020 through 2024. This projection assumes that, in the absence of the pandemic, employment would have continued to grow at a similar rate. The resulting trend is visualized in graph 6, which extends the pre-COVID trajectory while deliberately excluding the disruptive effects of the pandemic on the labour market.

![Graph 6](Graph_pngs/graph6.png)
##### Graph 6


In contrast to employment, population graph 7 was not extrapolated. Instead, observed population data for each year were used, as demographic growth was only marginally affected by COVID-19 and remained relatively stable.

![Graph 7](Graph_pngs/graph7.png)
##### Graph 7


Using these two inputs, projected employment and observed population, a linear regression model was trained on historical public transport usage data from 2014 to 2019. The model identifies how public transport usage (measured in travel distance and travel time) responds to changes in employment and population. The model was then applied to the post-2019 period to estimate public transport usage for 2020-2024, representing a scenario in which the pandemic had not occurred. The predicted values and 95% confidence intervals are presented in graph 8.

![Graph 8](Graph_pngs/graph8.png)
##### Graph 8


This approach enables a simulation of a hypothetical scenario in which public transport usage evolves as if COVID-19 pandemic had not occurred, providing a baseline for comparison with observed data. Each subplot shows predicted values with a 95% confidence interval (shaded area), indicating the uncertainty surrounding the yearly estimates. The R² value reflects the extent to which the model explains variation in public transport usage based on employment and population. Higher R² values indicate that the model’s predictions align more closely with the observed data, confirming that the selected input variables are strong predictors for  public transport usage.

The model fits train usage more closely than bus or metro usage, as evidenced by both higher R² values and narrower confidence intervals. This suggests that train commuting patterns are more consistently associated with employment and population trends, whereas bus and metro usage may be influenced by additional factors such as service changes, urban form, or modal competition.

A comparison between the model’s projections and the actual observed values, as shown in graph 9, illustrates the impact of the COVID-19 pandemic. For both train and bus/metro usage, the observed data from 2020 onwards show significant disruptions, with a sharp decline in travel distance in 2020 followed by a gradual recovery in subsequent years. The model’s predictions for 2020-2024, while based on pre-pandemic trends, do not fully capture this abrupt change, particularly in the first years. The actual values indicate that train and bus/metro usage were much lower in 2020 than predicted, but began to recover faster than the model anticipated. This highlights that, although the model provides a useful baseline, it does not account for the extraordinary disruptions caused by the pandemic. The recovery patterns for both modes are influenced by factors outside the scope of the model’s original predictors.

It should be noted that graph 9 specifically presents the total travel distance per day for both transport modes, but that trends in travel time follow a similar pattern. The discrepancies between predicted and actual values for travel distance are mirrored in the travel time data, further emphasizing the strong impact of the COVID-19 pandemic on both train and bus/metro usage.

## 3. Analysis on How the Shift to Hybrid Working Has Influenced Peak-Hour vs. Off-Peak Demand Across Weekdays

To examine how hybrid working has affected peak-hour and off-peak demand for public transport, data from the OV-monitor of CBR, derived from Translink, were analysed. The datasets contain the total number of check-ins across all public transport modes on working days only, aggregated by hour of the day and by weekday. Focusing on working days helps to isolate commuting patterns, as weekend travel typically reflects leisure rather than work-related mobility.

![Graph 10](Graph_pngs/graph10.png)
##### Graph 10


Graph 10 illustrates how hourly check-in patterns have evolved between 2019 (pre-COVID), 2021 (mid-COVID), and 2024 (post-COVID). Before the pandemic, in 2019, the data show two pronounced peaks, around 426,000 check-ins at 8:00 and 405,000 at 17:00, representing a sharply defined commuting rhythm with clear separation between work and leisure hours. During the pandemic in 2021, this pattern changed dramatically. Both peaks dropped by more than half, to roughly 192,000 and 193,000 check-ins, respectively. The morning rush hour was particularly affected, showing the strongest relative decline. This suggests that remote working replaced a large share of traditional morning commutes, while a smaller share of travellers still moved during later hours, likely for essential or flexible work purposes.

By 2024, peak-hour activity had partially recovered to 354,000 (8:00) and 346,000 (17:00) check-ins. However, compared to 2019, peak levels remained 15-20% lower, and the gap between the morning and evening peaks had narrowed. In addition, off-peak hours, especially between 10:00 and 15:00, saw noticeably higher check-in levels than in 2019, indicating a more even distribution of travel across the day.

Overall, the comparison across years shows a clear transition from concentrated rush-hour demand in 2019, to a drastically reduced and flattened pattern in 2021, and finally to a partially recovered but more spread-out pattern in 2024. This progression highlights how hybrid working not only reduced total commuting volumes but also blurred the temporal boundaries of traditional peak hours.


![Graph 11](Graph_pngs/graph11.png)
##### Graph 11

Graph 11 compares the average number of check-ins per weekday across the same three years, further illustrating how hybrid working reshaped commuting behaviour throughout the week. 

In 2019, weekday volumes were stable and high, fluctuating only slightly between 4.01 million on Monday and 4.32 million on Thursday, reflecting the traditional full-week office schedule. 

In 2021, average check-ins dropped uniformly to around 2.0-2.3 million per day, showing that the reduction in mobility affected all weekdays equally, a pattern consistent with nationwide lockdowns and mandatory remote work policies. 

By 2024, total volumes had increased again to roughly 3.5-4.0 million check-ins per day, yet the distribution across weekdays changed significantly. While Tuesday (3.83M) and Thursday (3.95M) became the busiest commuting days, Monday (3.48M) and Friday (3.63M) remained clearly lower. Compared with 2019, this indicates not just a partial recovery in overall demand, but a shift in the weekly rhythm of travel: midweek days regained much of their pre-COVID intensity, while the start and end of the week stayed quieter. 

This pattern aligns with findings from Employers’s association AWVN (Van der Velde, 2024), which report that 94% of Dutch employers now allow remote work, and that Tuesday and Thursday have become the most popular office days, while many employees work from home on Monday and Friday. These trends help explain why public transport demand in 2024 is highest midweek and lower at the start and end of the week, reflecting the influence of hybrid work schedules on commuting behaviour.

Together, Graphs 10 and 11 show that hybrid working has led to a dual shift: a flattening of hourly peaks and a redistribution of weekday demand. Commuting is now less synchronised in time and less uniform across the week, reflecting a lasting structural change in how and when people travel for work. ## 4. Conclusion

This research examined how the rise of hybrid working since the COVID-19 pandemic has influenced the demand for public transport in the Netherlands. By analysing national mobility data and public transport check-in statistics, the study identified clear and lasting changes in travel behaviour across modes, time periods, and weekdays.

The first analysis showed that all transport modes experienced a decline in commuting activity during the pandemic, but the impact was most severe for public transport. Train and bus/metro usage dropped to less than half of their pre-pandemic levels, while car use proved far more resilient. Although partial recovery occurred after 2021, public transport demand has not returned to its 2019 levels, indicating a structural rather than temporary shift in commuting behaviour.

The second analysis, which modelled a counterfactual scenario without the pandemic, demonstrated that public transport use today would have been considerably higher if previous growth trends had continued. Employment and population data suggest that, without COVID-19, demand for trains and buses would have followed a steady upward trajectory. The gap between this predicted trend and the observed data illustrates the long-term suppressing effect of the pandemic and the subsequent rise of hybrid working on public transport demand.

Finally, the analysis of hourly and weekday check-ins revealed that hybrid working has reshaped not only the volume but also the timing of travel. Peak-hour demand has flattened, and midweek travel has become more dominant, with Monday and Friday now significantly quieter. This reflects greater flexibility in work patterns and a reduced need for daily commuting.

In conclusion, hybrid working has led to a sustained reduction and redistribution of public transport demand in the Netherlands. Travel behaviour has become less concentrated in time and more dispersed across the week, signalling a lasting transformation in how people travel for work in the post-pandemic era.


## 5. Discussion / Limitations

The findings of this research indicate that the rise of hybrid working has structurally reshaped commuting behaviour and public transport demand in the Netherlands. Rather than representing a temporary disruption, the observed changes suggest a longer-term shift toward greater flexibility in travel timing and frequency. This aligns with broader societal trends toward digitalisation and flexible work arrangements, which have reduced the dominance of fixed office hours and regular commuting routines. 

From a theoretical perspective, these findings challenge traditional transport demand models that rely on stable, cyclical commuting behaviour. The results point to a more dynamic mobility landscape in which travel demand is less tied to specific hours or weekdays. Future modelling efforts should therefore account for hybrid work as a structural determinant of demand, not as an external disturbance. 

For policymakers and transport operators, the implications are substantial. Flatter and less predictable peak-hour patterns require more adaptive service planning, dynamic pricing strategies, and a reconsideration of capacity investments. Encouraging off-peak travel and integrating flexible ticketing options could help maintain efficiency while aligning with evolving commuter preferences. 

However, some limitations must be acknowledged. The analysis is based on aggregated CBS and Translink data, which exclude certain ticket types such as disposable cards and barcodes, possibly underestimating total check-ins. Moreover, the study focuses on national averages, overlooking potential spatial differences between urban and rural regions or between the Randstad and other areas. Data on the exact proportion of hybrid workers were also limited, constraining the direct linkage between work practices and mobility outcomes. 

Another limitation concerns the time horizon of the data. Most datasets extend only to 2024, which is roughly two years after the pandemic’s end. It remains uncertain whether the observed decline in public transport use will persist or continue to recover in the longer term. Future research should therefore monitor subsequent years to assess whether these trends represent a permanent transformation or a delayed adjustment phase.
